```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
and_op = pd.DataFrame()
or_op = pd.DataFrame()
xor_op = pd.DataFrame()
```

```python
and_op['input1'] =[ 1,1,0,0]
and_op['input2'] = [1,0,1,0]
and_op['output' ]=[1,0,0,0]
#
```

```python
or_op['input_1']=[ 1,1,0,0]
or_op['input_2']= [1,0,1,0]
or_op['output']= [1,1,1,0]
#
```

```python

xor_op['input_1']=[ 1,1,0,0]
xor_op['input_2']= [1,0,1,0]
xor_op['output']= [0,1,1,1]
#
```

```python
and_op
```

```python
sns.scatterplot(x='input1',y='input2',data=and_op)
```

```python
sns.scatterplot(x='input_1',y='input_2',data=or_op)
```

```python
xor_op
```

```python
sns.scatterplot(x='input_1', y= 'input_2', data =xor_op)
```

```python
from sklearn.linear_model import Perceptron
```

```python
cl1 = Perceptron()
cl2 = Perceptron()
cl3 = Perceptron()
```

```python
cl1.fit(and_op[['input1','input2']],and_op['output'])
cl2.fit(or_op[['input_1','input_2']],or_op['output'])
cl3.fit(xor_op[['input_1','input_2']],xor_op['output'])
```

```python
cl1.coef_
```

```python
cl1.intercept_
```
